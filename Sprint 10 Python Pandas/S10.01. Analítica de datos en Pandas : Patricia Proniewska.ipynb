```python
# creamos columna "Data de naixement" en formato datetime:

df["Data de naixement"] = pd.to_datetime(
    {"year": df["Any"],
    "month": df["Mes"],
    "day": df["Dia"]
    })

df.head(5)
```

```python
# definimos funcion que calcule la edad:

def calculadora_edad(fecha_nacimiento):
    actual = date.today()
    edad = actual.year - fecha_nacimiento.year
    edad = edad - ((actual.month, actual.day) < (fecha_nacimiento.month, fecha_nacimiento.day))
    return edad

# creamos la columna Edad:

df["Edat"] = df["Data de naixement"].apply(calculadora_edad)

df.head(5)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def crear_grafico(data_frame):
    for columna in data_frame.columns:

        if pd.api.types.is_numeric_dtype(data_frame[columna]):
            data_frame[columna].plot.hist(bins=30, title=columna)
            plt.title(columna)
            plt.savefig(f"{columna}.png")   # ⬅️ antes de show
            plt.show()
            plt.close()
            


        elif pd.api.types.is_datetime64_any_dtype(data_frame[columna]):
            data_frame[columna].dt.year.value_counts().plot.bar()
            plt.title(columna)
            plt.savefig(f"{columna}.png")   # ⬅️ antes de show
            plt.show()
            plt.close()
            

  
        else:
            data_frame[columna].value_counts().plot.bar()
            plt.title(columna)
            plt.savefig(f"{columna}.png")   # ⬅️ antes de show
            plt.show()
            plt.close()
            


```

">	2.
Carga el archivo matriz_distancies.xlsx en pandas, de modo que los nombres de filas y los nombres de columnas sean los de las ciudades. Borra "Las Palmas de Gran Canaria" y "Palma" para que podamos realizar el trayecto en coche.

Nos interesa visitar todas las principales ciudades de España recorriendo la mínima distancia posible.

No hace falta que lo hagas de forma óptima, nos interesa que desarrolles una solución razonable utilizando las herramientas que tienes actualmente.

Por ejemplo, una aproximación sencilla (que no óptima) sería yendo siempre a la ciudad más cercana que no hayamos visitado todavía

Haz una función que dada la matriz de distancias y la ciudad de origen, haga una propuesta de ruta que sea lo más corta posible que puedas, devolviendo una lista con el orden de visita. Da también la distancia total recorrida.

    
</div>

```python
import numpy as np
```

```python
df_distancias = pd.read_excel("matriu_distancies.xlsx", index_col=0)

df_distancias
```

```python
df_distancias = df_distancias.drop(columns=(["Palma", "Las Palmas de Gran Canaria"]))

```

```python
df_distancias = df_distancias.drop(index=["Palma", "Las Palmas de Gran Canaria"])
```

```python
df_distancias
```

```python
df_distancias.info()
```

```python
df_distancias = df_distancias.apply(pd.to_numeric, errors="coerce")

#Intenta convertir todas las columnas a números.
#Si algo no se puede convertir, pon NaN.”
```

```python
dist = df_distancias.copy()

#trabajaremos con dist para no tocar el Data Frame original.
```

```python
def crear_ruta_mas_corta(matriz_distancias, origen):

    dist = matriz_distancias.copy()
    np.fill_diagonal(dist.values, np.inf)

    actual = origen
    visitadas = [origen]
    distancia_total = 0.0

    ciudades = list(dist.index) #todas las ciudades que hay que visitar

    while len(visitadas) < len(ciudades): # seguimos MIENTRAS quede alguna ciudad para visitar (bucle principal WHILE)

        # Creamos una lista con las ciudades c que estan en ciudades pero que no estan en visitadas.
        no_visitadas = [c for c in ciudades if c not in visitadas]

        # distancias desde la ciudad actual a las no visitadas
        distancias_actual = dist.loc[actual, no_visitadas]

        # elegimos la ciudad más cercana
        siguiente = distancias_actual.idxmin() # idxmin dice el índice (nombre) del valor mas pequeño
        distancia = distancias_actual.loc[siguiente]

        # actualizamos estado
        distancia_total += float(distancia)
        visitadas.append(siguiente)
        actual = siguiente

    return visitadas, distancia_total


```

```python
ruta, total = crear_ruta_mas_corta(df_distancias, "Barcelona")
print(" → ".join(ruta))
print(total)

```

(EXTRA) Desde qué ciudad la ruta sería más corta con el algoritmo planteado

```python
# el algoritmo prueba el recorrido comenzando desde cada ciudad 
# para determinar desde cual se obtiene la menor distancia total 

# probar todas las ciudades como origen y quedarnos con la mejor (menor distancia total)

def conseguir_mejor_origen(matriz_distancias):

    mejor_origen = None                 # ciudad ganadora
    mejor_total = float("inf")          # empezamos con "infinito" para disminurir/mejorar la distancia
    mejor_ruta = None                  

    # probamos cada ciudad como punto de inicio
    for ciudad in matriz_distancias.index:
        ruta, total = crear_ruta_mas_corta(matriz_distancias, ciudad)  # usamos funcion creada antes

        # si esta ruta es más corta, la guardamos como la mejor
        if total < mejor_total:
            mejor_total = total
            mejor_origen = ciudad
            mejor_ruta = ruta

    return mejor_origen, mejor_total, mejor_ruta


# ---- CÓMO USARLO ----
mejor_origen, mejor_total, mejor_ruta = conseguir_mejor_origen(df_distancias)

print("Mejor ciudad de origen:", mejor_origen)
print("Distancia total:", mejor_total)
print("Ruta:")
print(" → ".join(mejor_ruta))
```
