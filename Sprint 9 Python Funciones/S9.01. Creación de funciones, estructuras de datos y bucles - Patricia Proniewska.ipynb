"> 1. Calculadora del √≠ndice de masa corporal

Escribe una funci√≥n que calcule el IMC ingresado por el usuario/a, es decir, quien lo ejecute deber√° ingresar estos datos. Puedes obtener m√°s informaci√≥n de su c√°lculo en: üîó √çndice de masa corporal IMC que es y c√≥mo se calcula.

La funci√≥n debe clasificar el resultado en sus respectivas categor√≠as.

Consejo: Intenta validar los datos previamente, para que env√≠e un mensaje de advertencia si los datos introducidos por el usuario/a no est√°n en el formato adecuado o no toma valores razonables.
    
</div>

```python
def calculadora_indice_masa_corporal():

    peso_usuario = float(input("1. Introduce tu peso (kilogramos): "))
    altura_usuario = float(input("2. Introduce tu tu altura en metros (usa punto, ej. 1.65): "))

    if peso_usuario <= 0 or altura_usuario <= 0:
        print(f"Error de datos")
        return
    
    if altura_usuario > 3:  
        print("Altura mal introducida. Usa metros, por ejemplo 1.65.")
        
        return

    imc = peso_usuario / (altura_usuario ** 2)


    if imc < 18.5:
        print(f"Tu IMC es {imc:.1f}. Categor√≠a: Bajo peso.")
    elif imc < 25:
        print(f"Tu IMC es {imc:.1f}. Categor√≠a: Peso normal.")
    elif imc < 30:
        print(f"Tu IMC es {imc:.1f}. Categor√≠a: Sobrepeso.")
    else:
        print(f"Tu IMC es {imc:.1f}. Categor√≠a: Obesidad.")
    

calculadora_indice_masa_corporal()
```

">2. Conversor de temperaturas

Existen diversas unidades de temperatura utilizadas en distintos contextos y regiones. Las m√°s comunes son Celsius (¬∞C), Fahrenheit (¬∞F) y Kelvin (K). Tambi√©n existen otras unidades como Rankine (¬∞Ra) y R√©aumur (¬∞Re).

Selecciona al menos 2 conversores, de modo que al introducir una temperatura devuelva, como m√≠nimo, dos conversiones, de modo que se puedan guardar (recuerda que un print() no se puede guardar nunca).

Consejo: Intenta validar los datos previamente, para que env√≠e un mensaje de advertencia si los datos introducidos por el usuario/a no est√°n en el formato adecuado.

(EXTRA): Piensa una manera de almacenar todas las posibles conversiones en un solo objeto (¬øLista? ¬øDiccionario? ¬øDataFrame?) en vez de escribir muchos if else en funci√≥n de la temperatura de origen y la temperatura de destino.

```python
def conversor_de_temperaturas():
    try: # # Intentamos convertir el input a un numero
        temperatura_celsius = float(input("1. Introduce la temperatura en Celsius (entero o decimal, ej. 21.5): "))
    except: # si algo sale mal, mostramos advertencia y salimos del programa
        print("Temperatura mal introducida. Por favor introduce un n√∫mero v√°lido.")
        return

    temperatura_fahrenheit = (temperatura_celsius * 9/5) + 32
    temperatura_kelvin = temperatura_celsius + 273.15

    resultados = {
        "Celsius (¬∞C)": temperatura_celsius,
        "Fahrenheit (¬∞F)": temperatura_fahrenheit,
        "Kelvin (K)": temperatura_kelvin}

    return resultados


mostrar_conversiones = conversor_de_temperaturas()
for clave, valor in mostrar_conversiones.items():
    print(clave, ":", valor)


```

">3. Contador de palabras de un texto.

Escribe una funci√≥n que dado un texto, muestre las veces que aparece cada palabra. Intenta que se gestionen todas las posibles casu√≠sticas que hagan que el programa no funcione correctamente.

(EXTRA): ¬øCu√°l es la longitud media de las palabras del texto que has escrito? ‚Äú¬øHola c√≥mo va?‚Äù deber√≠a devolver (4+3+2) / 3 = 3
    
</div>

```python
def contador_palabras():
    texto = input("Escribe un texto: ")
    print(texto)

    texto = texto.lower()

    signos = ".,;:!?¬ø¬°()\"'‚Äî-"
    for s in signos:
        texto = texto.replace(s, "")


    lista_palabras = texto.split()

    longitudes = []
    for palabra in lista_palabras:
        longitudes.append(len(palabra))
    media = sum(longitudes) / len(longitudes)


    contar = {}
    for palabra in lista_palabras:
        if palabra in contar:
            contar[palabra] += 1
        else:
            contar[palabra] = 1
    
    for palabra, veces in contar.items():
        print(palabra, ":", veces)

    print("La longitud media de las palabras es:", round(media, 2))

contador_palabras()
```

">4. Diccionario inverso (con posibilidad de duplicados)

Resulta que el cliente tiene una encuesta muy antigua que se almacena en un diccionario y los resultados los necesita a la inversa, es decir, intercambiando las claves y valores. Los valores y claves en el diccionario original son √∫nicos; si √©ste no es el caso, la funci√≥n deber√≠a imprimir un mensaje de advertencia, junto con una lista con los valores asociados a la clave repetida.
    
</div>

```python

def diccionario_inverso(diccionario):
    inverso = {}

    for clave, valor in diccionario.items():
        if valor not in inverso:
            inverso[valor] = [clave]      # primera vez que aparece - lista con la clave
        else:
            inverso[valor].append(clave)  # valor repetido - a√±adimos clave
            print("Advertencia: valor repetido:", valor)

    return inverso


peliculas = {
    1994: "Pulp Fiction",
    1999: "Matrix",
    2001: "Amelie",
    2003: "Matrix"
}

# Obtener diccionario invertido - llamar la funcion
resultado = diccionario_inverso(peliculas)

# Imprimir una linea por cada palabra
for titulo, a√±os in resultado.items():
    print(f"{titulo} : {a√±os}")
```

">1. Contador y ordenador de palabras de un texto.

El cliente qued√≥ contento con el contador de palabras, pero ahora quiere leer archivos TXT y que calcule la frecuencia de cada palabra ordenadas dentro de las entradas habituales del diccionario seg√∫n la letra con la que comienzan, es decir, las claves deben ir de la A a la Z y dentro de la A debemos ir de la A a Z. "tu_me_quieres_blanca.txt" la salida esperada ser√≠a:
    
</div>

```python
from pprint import pprint # para imprimir un diccionario bonito

def contador_ordenador_palabras():

    # 1. Leer el archivo completo
    with open("tu_me_quieres_blanca.txt", "r", encoding="utf-8") as archivo:
        texto = archivo.read()

    # 2. Reutilizamos el codigo de limpieza de signos de Nivel 1
    # Pasar a minusculas y limpio signos

    texto = texto.lower()

    signos = ".,;:!?¬ø¬°()\"'‚Äî-"
    for s in signos:
        texto = texto.replace(s, "")

    # 3. Creamos lista de palabras
    lista_palabras = texto.split()

    # 4. Creamos diccionario principal de letras
    resultado = {}

    # 5. Rellenamos diccionario - agrupando por letra inicial
    for palabra in lista_palabras:
        letra = palabra[0]  # primera letra de la palabra

        # Si la letra no existe aun, la creamos
        if letra not in resultado:
            resultado[letra] = {}

        # A√±adir o aumentar contador de esta palabra
        if palabra not in resultado[letra]:
            resultado[letra][palabra] = 1
        else:
            resultado[letra][palabra] += 1

    # 6. Ordenar letras (A‚ÄìZ) y palabras dentro de cada letra
    resultado_ordenado = {}

    for letra in sorted(resultado.keys()):
        # ordenar palabras dentro de cada letra
        palabras_ordenadas = dict(sorted(resultado[letra].items()))
        resultado_ordenado[letra] = palabras_ordenadas

    # 7. Imprimir el diccionario final ordenado
    pprint(resultado_ordenado)



# LLAMAR A LA FUNCI√ìN
contador_ordenador_palabras()
```

```python
datos_cliente = ['1.3', 'one', '1e10', 'seven', '3-1/2', ('2', 1, 1.4, 'not-a-number'), [1, 2, '3', '3.4']]

def conversion(lista):
    convertidos = []
    no_convertidos = []

    for elemento in lista:

        # 1: si es lista o tupla - recorremos este contenido (isinstance)
        if isinstance(elemento, (list, tuple)):
            for dato in elemento:   # nombre m√°s claro
                try:
                    convertidos.append(float(dato)) # intenta hacer esto
                except:
                    no_convertidos.append(dato) # si algo sale mal, haz esto otro

        # 2: elemento normal - intentar convertir directamente al float
        else:
            try:
                convertidos.append(float(elemento))
            except:
                no_convertidos.append(elemento)

    return convertidos, no_convertidos

resultado_convertidos, resultado_no_convertidos = conversion(datos_cliente)

print("Convertidos:", resultado_convertidos)
print("No convertidos:", resultado_no_convertidos)

# Usamos try/except para intentar convertir cada dato a float.
# Si la conversi√≥n funciona, lo guardamos en la lista de convertidos.
# Si falla, no se rompe el programa y guardamos el dato en no_convertidos.

```

Explora el funcionamiento del m√≥dulo random de la librer√≠a numpy. üîó Random module in NumPy

En este punto, el cliente ha detectado un problema con las contrase√±as que utilizan sus trabajadores. Asdf1234, fechas de cumplea√±os o similares. Para solucionarlo, nos ha encargado una funci√≥n de Python que genere contrase√±as m√°s seguras. La funci√≥n debe depender de los siguientes par√°metros:

longitud (int): Longitud de la contrase√±a

(EXTRA) Explora c√≥mo podr√≠amos hacer que la funci√≥n copiara la contrase√±a autom√°ticamente en el portapapeles del ordenador (como si la hubi√©ramos seleccionado y hecho ctrl+copy).

```python
import pyperclip
import numpy as np


def crear_contrase√±a(longitud, mayusculas=True, minusculas=True, numeros=True, signos=False):
    
    caracteres = []

    # A√±adimos grupos de caracteres segun los parametros
    if mayusculas:
        caracteres.extend(list("ABCDEFGHIJKLMNOPQRSTUVWXYZ"))
    if minusculas:
        caracteres.extend(list("abcdefghijklmnopqrstuvwxyz"))
    if numeros:
        caracteres.extend(list("0123456789"))
    if signos:
        caracteres.extend(list("!@#$%&*?"))

    # Comprobamos que hay caracteres disponibles
    if not caracteres:
        return "No se ha seleccionado ning√∫n tipo de car√°cter. Por favor, int√©ntalo de nuevo."


    # Generamos la contrase√±a con caracteres aleatorios
    contrase√±a = "".join(
        np.random.choice(caracteres, size=longitud)
    )  
    
    # Se seleccionan caracteres aleatorios de la lista y 
    # se unen para formar la contrase√±a. 
    # np.random.choice es una funci√≥n de NumPy 
    # que elige valores al azar de una lista

    pyperclip.copy(contrase√±a) # copiar al clipboard
    
    return contrase√±a

print(crear_contrase√±a(10, True, True, True, True))

print(crear_contrase√±a(7,False, True, True, False))

print(crear_contrase√±a(5,False, False, False, False))
```

"> 2. Procesamiento de datos simple

Una compa√±era de trabajo nos ha pedido un favor, aprovechando que sabe que estamos aprendiendo a programar. Tiene un hist√≥rico de partidos de f√∫tbol catal√°n en un fichero, donde se almacenan los nombres de los equipos y los resultados. Necesita que procesemos los datos de forma autom√°tica, para extraer los resultados que necesita.

Utiliza el archivo "historic_partits.txt"

Necesita un programa que devuelva:

El n√∫mero total de goles que ha marcado cada equipo.
El nombre del equipo m√°s goleador.
El nombre del equipo m√°s goleado
La clasificaci√≥n global (cada victoria: 3 pts, empate 1 pts, derrota 0 pts)
    
</div>

<div style="
    background-color:#cfe6ff;
    color:#000000;
    padding:16px;
    margin-bottom:16px;
"> DEFINIMOS LA FUNCION:
</div>

```python
def procesar_partidos():

    with open("historic_partits.txt", "r", encoding="utf-8") as archivo_partidos:
        
        texto = archivo_partidos.read()

        lineas = texto.splitlines() 
        # Separa el texto en l√≠neas, para que cada l√≠nea represente un partido

        partidos = []

        for linea in lineas:
            
            partes = linea.split("\t")
            # Separa cada l√≠nea por tabuladores: equipo 1 / resultado / equipo 2
            
            equipo_1 = partes[0]  # nombre de equipo 1
            goles = partes[1].split("-")  # separa goles equipo 1 y 2
            equipo_2 = partes[2] # nombre equipo 2

            goles_1 = int(goles[0]) # Goles del equipo 1
            goles_2 = int(goles[1]) # Goles del equipo 2

            partidos.append([equipo_1, goles_1, goles_2, equipo_2])
            # partidos ser√° una lista de listas

    return partidos
```

<div style="
    background-color:#cfe6ff;
    color:#000000;
    padding:16px;
    margin-bottom:16px;
"> PROCESAMOS LOS DATOS:
El n√∫mero total de goles que ha marcado cada equipo.
    
</div>

```python

partidos = procesar_partidos()

goles_por_equipo = {}

for partido in partidos:
    equipo_1, goles_1, goles_2, equipo_2 = partido

    # Inicializa el equipo 1 si no existe
    if equipo_1 not in goles_por_equipo:
        goles_por_equipo[equipo_1] = {"goles_marcados": 0, "goles_recibidos": 0}

    # Inicializa el equipo 2 si no existe
    if equipo_2 not in goles_por_equipo:
        goles_por_equipo[equipo_2] = {"goles_marcados": 0, "goles_recibidos": 0}

    # Suma goles marcados y recibidos
    goles_por_equipo[equipo_1]["goles_marcados"] += goles_1
    goles_por_equipo[equipo_1]["goles_recibidos"] += goles_2

    goles_por_equipo[equipo_2]["goles_marcados"] += goles_2
    goles_por_equipo[equipo_2]["goles_recibidos"] += goles_1

print(goles_por_equipo)
```

<div style="
    background-color:#cfe6ff;
    color:#000000;

    margin-bottom:16px;
"> El nombre del equipo m√°s goleador
    
</div>

```python

equipo_mas_goleador = ""
max_goles = 0

for equipo, goles in goles_por_equipo.items():
    # Compara los goles marcados de cada equipo
    if goles["goles_marcados"] > max_goles:
        max_goles = goles["goles_marcados"]
        equipo_mas_goleador = equipo

print("Equipo m√°s goleador:", equipo_mas_goleador)

```

```python
# Obtiene el nombre del equipo que ha recibido m√°s goles

equipo_mas_goleado = ""
max_goles_recibidos = 0

for equipo, goles in goles_por_equipo.items():
    # Compara los goles recibidos de cada equipo
    if goles["goles_recibidos"] > max_goles_recibidos:
        max_goles_recibidos = goles["goles_recibidos"]
        equipo_mas_goleado = equipo

print("Equipo m√°s goleado:", equipo_mas_goleado)

```

```python
# Calcula la clasificaci√≥n global por puntos:
#  Se recorren todos los partidos y se asignan puntos a cada equipo 
# seg√∫n el resultado: si ganaron, empate o perdida.

clasificacion = {}

for partido in partidos:
    equipo_1, goles_1, goles_2, equipo_2 = partido

    # Inicializa los equipos si no existen
    if equipo_1 not in clasificacion:
        clasificacion[equipo_1] = 0
    if equipo_2 not in clasificacion:
        clasificacion[equipo_2] = 0

    # Asigna puntos dependiendo del resultado
    # victoria = 3 puntos, empate = 1 punto, derrota = 0 puntos
    if goles_1 > goles_2:
        clasificacion[equipo_1] += 3
    elif goles_1 < goles_2:
        clasificacion[equipo_2] += 3

    # Si ninguno de los dos ha marcado mas goles,
    # significa que el partido ha terminado en empate
    # y cada equipo suma 1 punto
    else:
        clasificacion[equipo_1] += 1
        clasificacion[equipo_2] += 1

print(clasificacion)

# Ordenamos la clasificaci√≥n para tener el ranking:
clasificacion_ordenada = sorted(
    clasificacion.items(),
    key=lambda item: item[1],
    reverse=True
)

for equipo, puntos in clasificacion_ordenada:
    print(equipo, ":", puntos, "pts")


```
